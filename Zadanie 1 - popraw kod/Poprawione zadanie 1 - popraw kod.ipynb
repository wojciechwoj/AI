```python
def average_age(records):
    # SPRAWDZENIE BŁĘDU: Zwracamy 0 (lub np. None), jeśli lista jest pusta
    if not records:
        return 0
        
    total = 0
    for record in records:
        total += record['age']
        
    # Teraz bezpiecznie dzielimy, bo wiemy, że records ma długość > 0
    return total / len(records) 


if __name__ == '__main__':
    sample = [
        {'name': 'Anna', 'age': 21},
        {'name': 'Piotr', 'age': 23},
        {'name': 'Maria', 'age': 25},
    ]
    empty_sample = [] # Nasz testowy pusty przypadek
    
    print("Średnia dla próbki:", average_age(sample))
    print("Średnia dla pustej listy:", average_age(empty_sample))
```
