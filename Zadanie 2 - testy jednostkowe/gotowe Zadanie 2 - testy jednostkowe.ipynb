```python
def average_age(records):
    total = 0
    for record in records:
        total += record['age']
    # Dodano "s" na końcu - dzielimy przez długość listy "records"
    return total / len(records) 


if __name__ == '__main__':
    sample = [
        {'name': 'Anna', 'age': 21},
        {'name': 'Piotr', 'age': 23},
        {'name': 'Maria', 'age': 25},
    ]
    print(average_age(sample))
```

```python
import pytest
from risk_score import calculate_risk_score

# TEST 1: Standardowe, poprawne dane 
def test_calculate_risk_score_standard():
    # Wyliczenia: 50*0.2 (10) + 200*0.05 (10) - 150*0.03 (4.5) = 15.5
    assert calculate_risk_score(50, 200, 150) == 15.50

# TEST 2: Przypadek, w którym zaokrąglenie (round) odgrywa rolę
def test_calculate_risk_score_rounding():
    # 33*0.2 (6.6) + 181*0.05 (9.05) - 177*0.03 (5.31) = 10.34
    assert calculate_risk_score(33, 181, 177) == 10.34

# TEST 3: Oczekiwany błąd przy ujemnym wieku
def test_calculate_risk_score_negative_age():
    with pytest.raises(ValueError, match='All parameters must be positive.'):
        calculate_risk_score(-5, 200, 150)

# TEST 4: Oczekiwany błąd przy zerowym tętnie
def test_calculate_risk_score_zero_heart_rate():
    with pytest.raises(ValueError, match='All parameters must be positive.'):
        calculate_risk_score(40, 180, 0)
```

```python
# TEST 5: Edge Case - najmniejsze możliwe, ale wciąż legalne parametry (na granicy)
def test_calculate_risk_score_edge_case():
    # Wiek 0 i cholesterol 0 są dopuszczane przez funkcję. Najmniejsze tętno to np. 1.
    # Wyliczenia: 0*0.2 (0) + 0*0.05 (0) - 1*0.03 (0.03) = -0.03
    assert calculate_risk_score(0, 0, 1) == -0.03
```
