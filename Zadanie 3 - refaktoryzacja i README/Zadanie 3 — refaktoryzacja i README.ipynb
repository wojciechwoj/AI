```python
# risk_score.py

# Stałe wagi dla poszczególnych parametrów zdrowotnych
WEIGHT_AGE = 0.2
WEIGHT_CHOLESTEROL = 0.05
WEIGHT_HEART_RATE = 0.03

def calculate_risk_score(age: int, cholesterol: int, max_heart_rate: int) -> float:
    """
    Oblicza wskaźnik ryzyka sercowo-naczyniowego na podstawie wieku, cholesterolu i tętna.
    Wszystkie parametry wejściowe muszą być wartościami dodatnimi.
    """
    if age < 0 or cholesterol < 0 or max_heart_rate <= 0:
        raise ValueError('All parameters must be positive.')

    score = (age * WEIGHT_AGE) + (cholesterol * WEIGHT_CHOLESTEROL) - (max_heart_rate * WEIGHT_HEART_RATE)
    return round(score, 2)
```
